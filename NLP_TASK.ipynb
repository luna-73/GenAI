```python
import spacy

# load model
nlp = spacy.load("en_core_web_sm")

# -------- Example 1 --------
text1 = "Apple is looking at buying a startup in India for $1 billion. Elon Musk founded SpaceX."
doc1 = nlp(text1)

print("===== EXAMPLE 1 =====")

print("\n---- NAMED ENTITY RECOGNITION (NER) ----")
for ent in doc1.ents:
    print(ent.text, "->", ent.label_)

print("\n---- PART OF SPEECH (POS TAGGING) ----")
for token in doc1:
    print(token.text, "->", token.pos_)

print("\n---- CHUNKING (NOUN PHRASES) ----")
for chunk in doc1.noun_chunks:
    print(chunk.text)


# -------- Example 2 --------
text2 = "Jiya submitted her final project report at Mumbai University on Monday and received positive feedback from her professor."
doc2 = nlp(text2)

print("\n\n===== EXAMPLE 2 =====")

print("\n---- NAMED ENTITY RECOGNITION (NER) ----")
for ent in doc2.ents:
    print(ent.text, "->", ent.label_)

print("\n---- PART OF SPEECH (POS TAGGING) ----")
for token in doc2:
    print(token.text, "->", token.pos_)

print("\n---- CHUNKING (NOUN PHRASES) ----")
for chunk in doc2.noun_chunks:
    print(chunk.text)
```

```python
from spacy import displacy

print("Example 1 Visualization:")
displacy.render(doc1, style="ent", jupyter=True)

print("\nExample 2 Visualization:")
displacy.render(doc2, style="ent", jupyter=True)
```
